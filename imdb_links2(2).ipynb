```python
#import libraries
import pandas as pd
import numpy as np
import json
import requests
from bs4 import BeautifulSoup
```

```python
#enter api key and show name as variables
#enter show search url as variable
#search for show id 
api_key = "k_793904nk/" 
show = "Wynonna Earp" 
show_search_url = "https://imdb-api.com/en/API/SearchTitle/" + api_key + show
s = requests.get(show_search_url)
s.json()
#flatten s1
s1 = json.loads(s.text)
df = pd.json_normalize(s1, ['results'])
#print (df)
show_id = df.loc[df['title'] == show,'id'].item()
#print (show_id)
```

```python
#enter cast search url as variable
#search for cast members and ids
cast_search_url = "https://imdb-api.com/en/API/FullCast/" + api_key + show_id
c = requests.get(cast_search_url)
c.json()
#flatten c1
c1 = json.loads(c.text)
df2 = pd.json_normalize(c1, ['actors'])
df2.d
print (df2)
actor_ids = df2['id'].to_list()
print (actor_ids)
```
